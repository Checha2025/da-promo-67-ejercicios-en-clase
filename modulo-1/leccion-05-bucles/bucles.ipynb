```python
# adivina el numero
numero =30
numero_usuario = int(input("Adivina mi numero"))

while numero != numero_usuario:
    print("Has fallado, vuelve a introducir")
    numero_usuario = int(input(" Has fallado, adivina mi numero"))

if numero_usuario == numero: # >Tambein se puede usar un else
    print("Has ganado")
```
