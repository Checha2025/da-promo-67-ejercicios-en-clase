```python
# quiero tener una lista de numeros [1, 3,45, 789] y quiero que sume cada uno de ellos. 1+3, y ese resultado + 45

lista = [1, 3,45, 789]
suma_total = 0
for numero in lista: 
    suma_total = suma_total+numero # sumatotal = 0+1, sumatotal = 1+3, sumatotal = 4 + 45....

print(suma_total)
```

```python
def sumar_lista(lista):
    suma_total = 0
    for numero in lista: 
        suma_total = suma_total+numero # sumatotal = 0+1, sumatotal = 1+3, sumatotal = 4 + 45....
    return suma_total
```

```python
lista2 = [1, 3,45, 75, 78]
lista3 = [1, 3,45, 89, 78]
```

```python
sumar_lista(lista2)
```

```python
sumar_lista(lista3)
```

```python
lista_nombres =[" Nati ", " Rocio "," Ani ", " Ceci "]
lista_apellidos = [] 
```

```python


def limpiar(maricarmen):
    #necesito remover los espacios: strip()
    #necesito recorrer la lista: for
    # la lista se lo paso por parametros
    # retornar la lista limpia
    try:
        lista_vacia = []
        for elemento in maricarmen: 
            lista_vacia.append(elemento.strip().title())
        return lista_vacia
    except:
        print("Ha ocurrido un error")
```

```python
limpiar(maricarmen=lista_nombres) ## ejecutar la función
```

```python
limpiar(lista_apellidos)
```

```python
casa = 0
limpiar(casa)
```
