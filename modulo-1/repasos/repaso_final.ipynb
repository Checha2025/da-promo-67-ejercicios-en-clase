```python
# 1. estructuras
# libro = {isbn, titulo, autor, cantidad_total,cantidad_prestada}

INVENTARIO_LIBROS = []
USUARIOS = {}
MULTAS_COBRADAS = 0.0
ID_PRESTAMO_COUNTER = 1
```

```python
INVENTARIO_LIBROS = [{isbn, titulo, autor, cantidad_total,cantidad_prestada}, {isbn, titulo, autor, cantidad_total,cantidad_prestada}]
```

```python
# 1 . agregar_libro(isbn, titulo, autor, cantidad_total):
# Si el ISBN existe en INVENTARIO_LIBROS, aumenta su cantidad_total. Si no, agrega el nuevo libro con cantidad_prestada: 0.

def agregar_libro(isbn, titulo, autor, cantidad_total):
    for libro in INVENTARIO_LIBROS:
        if libro["isbn"] == isbn:
            libro["cantidad_total"] =  libro["cantidad_total"]+cantidad_total
            return "Stock actualizado"
    
    nuevo_libro = {
                'isbn' : isbn,
                'titulo': titulo,
                'autor': autor,
                'cantidad_total': cantidad_total,
                'cantidad_prestada': 0
            }   
    INVENTARIO_LIBROS.append(nuevo_libro)
    return f"Libro añadido {titulo}"
```

```python
isbn_nuevo = input("Dime el isbn")
titulo = input("Dime el titulo")
autor = input("Dime el autor")
cant = int(input("Dime la cantidad"))

agregar_libro(isbn_nuevo, titulo, autor, cant)
```

```python
#eliminar_libro(isbn):
# Busca y elimina el libro. Condición: Solo si la cantidad_prestada es $0$.

def eliminar_libro(isbn):
    ## código para eliminaar
    for libro in INVENTARIO_LIBROS:
        if libro["isbn"] == isbn :
            if libro['cantidad_prestada']==0:
                INVENTARIO_LIBROS.remove(libro)
                return "Libro eliminado correctamente"
            else:
                return "No puedo eliminar el libro porque tiene prestamos"
    return "Libro no encontrado, no lo puedo eliminar"
```

```python
libro_a_eliminar = input("Dime el isbn del libro a eliminar")
eliminar_libro(libro_a_eliminar)
```
